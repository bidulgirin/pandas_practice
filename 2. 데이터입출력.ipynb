- read_csv() 함수 옵션
    - path
    - 파일의 위치(파일명 포함) 또는 url

    - sep 도는 delimiter
        - 텍스트 데이터를 필드별로 구분하는 문자
        - 쉼표 대신 탭이나 공백으로 텍스트를 구분하는 경우 사용
    - header
        - 열 이름으로 사용될 행의 번호
        - 기본값은 0
        - header 가 없고 첫 행부터 데이터가 있는 경우 None 으로 지정
    - index_col
        - 행 인덱스로 사용할 열의 번호 또는 열 이름
    - names
        - 열 이름으로 사용할 문자열의 리스트
    - skiprows
        - 처음 몇 줄을 skip 할것인지 설정(숫자 입력)
        - skip 하려는 행의 번호를 담은 리스트로 설정 가능
    - parse_datas
        - 날짜 텍스트를 datetime64로 변환할 것인지 서정
        - 기본값은 False
    - skip_footer
        - 마지막 몇줄을 스킴할것인지 (숫자 입력)
     
    - encoding
        - 텍스트 인코딩 종류를 지정 ( 기본값은 utf-8 )

## 여러개의 데이터 프레임을 하나의 excel 파일로 저장

- 판다스 ExcelWriter() 함수는 excel 워크북 객체를 생성
- 워크북 객체는 일종의 excel 파일
- to_excel() 메소드를 적용할 때 삽입하려는 워크북 객체를 인자로 전달
- sheet_name 옵션에 excel 파일의 시트이름을 입력하여 삽입되는 시트 위치를 지정할 수 있음

```python
df
```

```python
data = {"c0" : [1,2,3],
       "c1" : [4,5,6],
        "c2" : [7,8,9]
       }

df2 = pd.DataFrame(data)
df2 = df2.set_index("c0")
```

```python
df2
```

```python
writer = pd.ExcelWriter("./df_excelwriter.xlsx") # 파일생성됨
```

```python
df.to_excel(writer, sheet_name="shit1")
df2.to_excel(writer, sheet_name = "shit2")
writer.close() # 꼭 닫아주자!!!!!!!
```
