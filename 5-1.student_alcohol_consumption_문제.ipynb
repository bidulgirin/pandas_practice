# 문제3. 문자열 데이터를 입력 받고 capitalize() 메소드를 적용하는 capitalizer 함수 정의

```python
def capitalizer(str):
    return str.capitalize()
```

# 문제4. Mjob과 Fjob열에 capitalizer 함수 적용된 결과 출력

```python
df["Mjob"] = df["Mjob"].map(lambda x : capitalizer(x) )
df["Fjob"] = df["Fjob"].map(lambda x : capitalizer(x) )

# 합쳐서 df[["Mjob", "Fjob"]].map(capitalizer) 이렇게 쓰는게 더 깔끔하겠죵~
df.head()
```
