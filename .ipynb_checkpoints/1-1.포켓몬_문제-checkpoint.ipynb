```python
import pandas as pd
```

```python
raw_data = {"이름": ['이상해씨', '파이리','꼬부기','캐터피'],
            "진화형": ['이상해풀','리자드','어니부기','단데기'],
            "타입": ['풀', '불', '물', '벌레'],
            "체력": [45, 39, 44, 45],
            "도감등록": ['yes', 'no','yes','no']                        
            }
```

# 문제1. raw_data를 데이터프레임 형태로 poke 변수에 할당

```python
poke = pd.DataFrame(raw_data)
```

```python
poke
```

# 문제2. 데이터프레임의 컬럼을 이름, 타입, 체력, 진화형, 도감등록 순으로 수정

```python
poke = poke[["이름","타입","체력","진화형","도감등록"]]

poke
```

- 파이리 : 도로
